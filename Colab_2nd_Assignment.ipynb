#### Necessary Libraries:

```python
import pandas as pd
from sklearn import model_selection
from sklearn import linear_model
```

### 1- Data Importing and Preprocessing

#### --> a- Reading the CSV File:

```python
breast_cancer_df = pd.read_csv("/Users/apple/Downloads/breast_cancer.csv")
breast_cancer_df.head()
```

#### --> b- Setting the index:

```python
# Use the id column as an index for the dataframe:
breast_cancer_df = breast_cancer_df.set_index(["id"])
breast_cancer_df.head()
```

#### --> c- Handling Missing Data:

```python
# find the overall number of missing values in the dataframe:
breast_cancer_df.isna().sum().sum()
```

```python
# Replace the missing values with the averages of the recspected columns:
breast_cancer_df.fillna(breast_cancer_df.mean(), inplace=True)
```

```python
# Check if we still have any missing data:
breast_cancer_df.isna().sum().sum()
```

#### --> d- Transform Diagnosis Variable from M & B to 1 & 0:

```python
# Function for defining a dictionary for a classified column:
def create_dict(column):
    values = column.unique()
    dictionary = {}
    dictionary_rev = {}
    for index, value in enumerate(values):
        dictionary[value] = index
        dictionary_rev[index] = value   
    return dictionary, dictionary_rev
```

```python
# Define a dictionary for diagnosis column to recode it:
diagnosis_dict, diagnosis_dict_rev = create_dict(breast_cancer_df.diagnosis)
```

```python
# Replace the diagnosis column with recoded values:
breast_cancer_df.diagnosis.replace(diagnosis_dict, inplace=True)
```

```python
# Check if the replacemnet process was successful:
breast_cancer_df.head()
```

### 2- Splitting the Data into Inputs/Features (x) & Output/Target (y)

#### --> a- Defining the Target & Features Variables:

```python
y = breast_cancer_df.diagnosis
x = breast_cancer_df.iloc[:,1:]
```

### --> b- Split the Variables into Training & Test Sets:

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y,test_size=0.1,random_state=101,stratify=y)
```

### 3- Define a Logistic Regression Model for Classifying the Breast Cancer Target Variable:

```python
breast_canser_logistic_model = linear_model.LogisticRegression(random_state=101)
```

### 4- Train the Logistic Regression Model:

```python
breast_canser_logistic_model.fit(x_train, y_train)
```

### 5- Evaluating the Accuracy of the Model:

```python
breast_canser_logistic_model.score(x_test,y_test)
```
